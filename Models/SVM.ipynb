```python
import numpy as np
import pandas as pd

# Path to your CSV file
csv_file_path = "Final_DF.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the DataFrame
df.info()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# Fit and transform the data
encoded_data = encoder.fit_transform(df['category'])
country = encoder.fit_transform(df['country'])
city = encoder.fit_transform(df['city'])
region = encoder.fit_transform(df['region'])
trx_type = encoder.fit_transform(df['trx_type'])
trx_status = encoder.fit_transform(df['trx_status'])
browsers = encoder.fit_transform(df['browser_environment'])
business = encoder.fit_transform(df['business_type_class'])
'''
# Create a dictionary to store the encoding information
encoding_info = {category: label for category, label in zip(data, encoded_data)}
'''

df['category'] = encoded_data
df['country'] = country
df['city'] = city
df['region'] = region
df['trx_status'] = trx_status
df['trx_type'] = trx_type
df['browser_environment'] = browsers
df['business_type_class'] = business
```
